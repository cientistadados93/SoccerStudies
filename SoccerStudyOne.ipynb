```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
brazilA = pd.read_csv('https://www.football-data.co.uk/new/BRA.csv')
premier = pd.read_csv('/content/drive/MyDrive/CSV/premierleague.csv')
laliga = pd.read_csv('/content/drive/MyDrive/CSV/laliga.csv')
serieA = pd.read_csv('/content/drive/MyDrive/CSV/serieA.csv')
ligue1 = pd.read_csv('/content/drive/MyDrive/CSV/ligue1.csv')
bundesliga = pd.read_csv('/content/drive/MyDrive/CSV/bundesliga.csv')
```

```python
leagues = ['Brazil', 'Premier', 'La Liga', 'Serie A', 'Ligue 1', 'Bundesliga']
home_win = [len(brazilA[brazilA.Res== 'H']), len(premier[premier.FTR == 'H']), len(laliga[laliga.FTR == 'H']), len(serieA[serieA.FTR== 'H']), len(ligue1[ligue1.FTR == 'H']), len(bundesliga[bundesliga.FTR == 'H'])]
draw = [len(brazilA[brazilA.Res== 'D']), len(premier[premier.FTR == 'D']), len(laliga[laliga.FTR == 'D']), len(serieA[serieA.FTR== 'D']), len(ligue1[ligue1.FTR == 'D']), len(bundesliga[bundesliga.FTR == 'D'])]
away_win = [len(brazilA[brazilA.Res== 'A']), len(premier[premier.FTR == 'A']), len(laliga[laliga.FTR == 'A']), len(serieA[serieA.FTR== 'A']), len(ligue1[ligue1.FTR == 'A']), len(bundesliga[bundesliga.FTR == 'A'])]
```

```python
x = np.arange(len(leagues)) 
plt.figure(figsize=(10,8)) 

plot = plt.bar(x, home_win, color=['#00FA9A', '#00FF7F', '#98FB98', '#90EE90', '#32CD32', '#228B22'])

for i in plot.patches:
  plt.annotate(i.get_height(),
                (i.get_x() + i.get_width()/2,i.get_height()),
                ha='center', va='baseline',fontsize=20,
                xytext=(0,3),
                textcoords='offset points'
                )
  
plt.title("Home wins in six leagues", fontsize="22")
plt.yticks(size='13')
plt.xticks(x, leagues, size='15')
plt.ylim(0,2000)
plt.ylabel('Number for wins', size='15')
plt.savefig("/content/drive/MyDrive/Images/homewins.jpeg")
plt.show()
```

```python
totalmatches = [len(brazilA), len(premier), len(laliga), len(serieA), len(ligue1), len(bundesliga)]
print(totalmatches)
```

```python
home_rate = [(float(home_win[0])/(3800))*100, (float(home_win[1])/(3800))*100, (float(home_win[2])/(3800))*100, (float(home_win[3])/(3800))*100, (float(home_win[4])/(totalmatches[4]))*100, (float(home_win[5])/(3060))*100]
```

```python
home_rate = [float('{:.2f}'.format(home_rate[0])), float('{:.2f}'.format(home_rate[1])), float('{:.2f}'.format(home_rate[2])), float('{:.2f}'.format(home_rate[3])), float('{:.2f}'.format(home_rate[4])), float('{:.2f}'.format(home_rate[5])) ]
print(home_rate)
```

```python
x = np.arange(len(leagues)) 
plt.figure(figsize=(10,8)) 

plot = plt.bar(x, home_rate, color=['#006400', '#00008B', '#FF0000', '#8B0000', '#8B008B', '#FFD700'])

for i in plot.patches:
  plt.annotate(i.get_height(),
                (i.get_x() + i.get_width()/2,i.get_height()),
                ha='center', va='baseline',fontsize=20,
                xytext=(0,3),
                textcoords='offset points'
                )
  
plt.title("Home wins in six leagues %", fontsize="22")
plt.yticks(size='13')
plt.xticks(x, leagues, size='15')
plt.ylim(0,60)
plt.ylabel('Home wins %', size='15')
plt.savefig("/content/drive/MyDrive/Images/homewinspercent.jpeg")
plt.show()
```
