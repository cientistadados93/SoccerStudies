### **Carregar Bibliotecas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### **Base de dados**

```python
brazilA = pd.read_csv('https://www.football-data.co.uk/new/BRA.csv')
premier = pd.read_csv('/content/drive/MyDrive/CSV/premierleague.csv')
laliga = pd.read_csv('/content/drive/MyDrive/CSV/laliga.csv')
serieA = pd.read_csv('/content/drive/MyDrive/CSV/serieA.csv')
ligue1 = pd.read_csv('/content/drive/MyDrive/CSV/ligue1.csv')
bundesliga = pd.read_csv('/content/drive/MyDrive/CSV/bundesliga.csv')

leagues = ['Brasileirão', 'Premier', 'La Liga', 'Serie A', 'Ligue 1', 'Bundesliga']
```

### **Analise das Vitórias de equipes da Casa e Visitante, e Empates**

```python
home_win = [len(brazilA[brazilA.Res== 'H']), len(premier[premier.FTR == 'H']), len(laliga[laliga.FTR == 'H']), len(serieA[serieA.FTR== 'H']), len(ligue1[ligue1.FTR == 'H']), len(bundesliga[bundesliga.FTR == 'H'])]
draw = [len(brazilA[brazilA.Res== 'D']), len(premier[premier.FTR == 'D']), len(laliga[laliga.FTR == 'D']), len(serieA[serieA.FTR== 'D']), len(ligue1[ligue1.FTR == 'D']), len(bundesliga[bundesliga.FTR == 'D'])]
away_win = [len(brazilA[brazilA.Res== 'A']), len(premier[premier.FTR == 'A']), len(laliga[laliga.FTR == 'A']), len(serieA[serieA.FTR== 'A']), len(ligue1[ligue1.FTR == 'A']), len(bundesliga[bundesliga.FTR == 'A'])]
```

```python
x = np.arange(len(leagues)) 
plt.figure(figsize=(10,8)) 

plot = plt.bar(x, home_win, color=['#12F150', '#12F152', '#12F154', '#12F156', '#12F158', '#12F160'])

for i in plot.patches:
  plt.annotate(i.get_height(),
                (i.get_x() + i.get_width()/2,i.get_height()),
                ha='center', va='baseline',fontsize=20,
                xytext=(0,3),
                textcoords='offset points'
                )
  
plt.title('Vitórias das equipes mandantes', fontsize='20')
plt.yticks(size='15')
plt.xticks(x, leagues, size='15')
plt.ylim(0,2000)
plt.ylabel('Número de vitórias', size='15')
plt.xlabel('Temporada 2011/12 até 2020/21', size='18')
plt.savefig("/content/drive/MyDrive/Images/homewins.jpeg")
plt.show()
```

### **Analise de Gols**

```python
gBrazil = pd.DataFrame(brazilA.HG + brazilA.AG)
gPremier = pd.DataFrame(premier.FTHG + premier.FTAG)
gLaliga = pd.DataFrame(laliga.FTHG + laliga.FTAG)
gSerieA = pd.DataFrame(serieA.FTHG + serieA.FTAG)
gLigue1 = pd.DataFrame(ligue1.FTHG + ligue1.FTAG)
gBundesliga = pd.DataFrame(bundesliga.FTHG + bundesliga.FTAG)
AllGoalssum = [int(gBrazil.sum()), int(gPremier.sum()), int(gLaliga.sum()),int(gSerieA.sum()), int(gLigue1.sum()), int(gBundesliga.sum())]
```

```python
y = leagues
x = AllGoalssum
```

```python
plt.figure(figsize=(15,12))
plot = plt.barh(y,x, color='#34c9eb')

for pY,pX in enumerate(AllGoalssum):
  plt.annotate(pX, xy=(pX,pY),fontsize=25, ha ='right', va = 'center')           

plt.title('Temporada 2011/12 até 2020/21', fontsize='20')
plt.yticks(size='18')
plt.xticks(size='18')
plt.xlim(0,11000)
plt.ylabel('Ligas', size='20')
plt.xlabel('Total de gols', size='18')
plt.savefig("/content/drive/MyDrive/Images/totalgoals.jpeg")
plt.show()
```

```python
goals_mean = [(float(AllGoalssum[0])/(3800)), (float(AllGoalssum[1])/(3800)), (float(AllGoalssum[2])/(3800)), (float(AllGoalssum[3])/(3800)), (float(AllGoalssum[4])/(3800)), (float(AllGoalssum[5])/(3060))]
goals_mean = [float('{:.2f}'.format(goals_mean[0])), float('{:.2f}'.format(goals_mean[1])), float('{:.2f}'.format(goals_mean[2])), float('{:.2f}'.format(goals_mean[3])), float('{:.2f}'.format(goals_mean[4])), float('{:.2f}'.format(goals_mean[5])) ]
print(goals_mean)
```

```python
x = goals_mean


plt.figure(figsize=(15,12))
plot = plt.barh(y,x, color='#b9fc00')

for pY,pX in enumerate(goals_mean):
  plt.annotate(pX, xy=(pX,pY),fontsize=25, ha ='right', va = 'center')           

plt.title('Temporada 2011/12 até 2020/21', fontsize='22')
plt.yticks(size='20')
plt.xticks(size='20')
plt.xlim(0,3)
plt.xlabel('Média de gols x partida', size='20')
plt.savefig("/content/drive/MyDrive/Images/totalgoalspermatch.jpeg")
plt.show()
```

```python
gHomeBrazil = pd.DataFrame(brazilA.HG)
gHomePremier = pd.DataFrame(premier.FTHG)
gHomeLaliga = pd.DataFrame(laliga.FTHG)
gHomeSerieA = pd.DataFrame(serieA.FTHG)
gHomeLigue1 = pd.DataFrame(ligue1.FTHG)
gHomeBundesliga = pd.DataFrame(bundesliga.FTHG)

gHomeTotals =[int(gHomeBrazil.sum()), int(gHomePremier.sum()), int(gHomeLaliga.sum()),int(gHomeSerieA.sum()), int(gHomeLigue1.sum()), int(gHomeBundesliga.sum())]

gHome_mean = [(float(gHomeTotals[0])/(3800)), (float(gHomeTotals[1])/(3800)), (float(gHomeTotals[2])/(3800)), (float(gHomeTotals[3])/(3800)), (float(gHomeTotals[4])/(3800)), (float(gHomeTotals[5])/(3060))]
gHome_mean = [float('{:.2f}'.format(gHome_mean[0])), float('{:.2f}'.format(gHome_mean[1])), float('{:.2f}'.format(gHome_mean[2])), float('{:.2f}'.format(gHome_mean[3])), float('{:.2f}'.format(gHome_mean[4])), float('{:.2f}'.format(gHome_mean[5])) ]

print(gHome_mean)
```

```python
gAwayBrazil = pd.DataFrame(brazilA.AG)
gAwayPremier = pd.DataFrame(premier.FTAG)
gAwayLaliga = pd.DataFrame(laliga.FTAG)
gAwaySerieA = pd.DataFrame(serieA.FTAG)
gAwayLigue1 = pd.DataFrame(ligue1.FTAG)
gAwayBundesliga = pd.DataFrame(bundesliga.FTAG)

gAwayTotals =[int(gAwayBrazil.sum()), int(gAwayPremier.sum()), int(gAwayLaliga.sum()),int(gAwaySerieA.sum()), int(gAwayLigue1.sum()), int(gAwayBundesliga.sum())]

gAway_mean = [(float(gAwayTotals[0])/(3800)), (float(gAwayTotals[1])/(3800)), (float(gAwayTotals[2])/(3800)), (float(gAwayTotals[3])/(3800)), (float(gAwayTotals[4])/(3800)), (float(gAwayTotals[5])/(3060))]
gAway_mean = [float('{:.2f}'.format(gAway_mean[0])), float('{:.2f}'.format(gAway_mean[1])), float('{:.2f}'.format(gAway_mean[2])), float('{:.2f}'.format(gAway_mean[3])), float('{:.2f}'.format(gAway_mean[4])), float('{:.2f}'.format(gAway_mean[5])) ]

print(gAway_mean)
```
